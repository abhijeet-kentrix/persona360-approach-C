```python
def get_income_group(income):
    if income >= 300000:
        return 5
    elif income < 300000 and income >= 230000:
        return 4
    elif income < 230000 and income >= 130000:
        return 3
    elif income < 130000 and income >= 80000:
        return 2
    elif income < 80000 and income >= 40000:
        return 1
    elif income < 40000:
        return 0
```

```python
df['Socio-Economic Profile : Income']=df['income_new'].apply(lambda x:get_income_group(x))
```
